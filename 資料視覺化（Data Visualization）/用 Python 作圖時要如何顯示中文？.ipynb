# 問題：使用 Python 作圖時中文字無法正常顯示，該怎麼辦？
示範環境：
- python 3.8.3
- matplotlib 3.3.2 
- seaborn 0.11.0

```python
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
# 調整作圖解析度
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 120
```

```python
ax = sns.barplot(x="day", y="total_bill", data=df)
```

```python
ax = sns.barplot(x="day", y="total_bill", data=df)
ax.set(xlabel='星期幾', ylabel='帳單總額')
```

# 解法：修改 matplotlib 預設字型
plt.rcParams['font.family']
##### 適用 matplotlib、pandas、seaborn 等

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family']
```

### 列出 matplotlib 可用的中文字型
註：需知道欲使用中文字型的英文名稱

```python
from matplotlib.font_manager import fontManager
[f.name for f in fontManager.ttflist]
```

### 修改 matplotlib 預設字型為欲使用的中文字型

```python
plt.rcParams['font.family'] = 'jf-openhuninn-1.0'
```

```python
ax = sns.barplot(x="day", y="total_bill", data=df)
ax.set(xlabel='星期幾', ylabel='帳單總額')
```
